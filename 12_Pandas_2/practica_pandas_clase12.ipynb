## Ejercicio 1.8 --> Merge dataframe

Lo que se busca con merge es unificar dos tablas seleccionando que queremos dejar. Es necesario que exista un campo en ambas

tablas que contenga la misma informacion, para el merge con dicho campo

Existen diferentes tipos de merge:

inner --> El cual solo deja los registros con los mismo valores del campo mergeado que se encuentras en ambas tablas

left --> Deja los que estan en comun, mas dotos los de la primer tabala

right --> Deja los que estan en comun mas todos los de la segunda tabla

outer --> deja todos los registros de ambas tablas, sin repetir



```python
df_empleados_cobro = pd.DataFrame({
    'ID': [101, 102, 103, 104,107,108,109],
    'Sueldo': [5000, 6000, 4500, 5500,6000,4000,4700],
    'Facturacion': [True, True, False, True,-False,False,True],
    'Antiguedad': [3, 5, 2, 8,1,7,6]
})

df_empleados_info = pd.DataFrame({
    'ID': [101, 102, 105, 106,107,109,110],
    'Edad': [31, 36, 29, 41,20,30,26],
    'Cargo': ['Analista', 'Gerente', 'Analista', 'Director','Administrativo','Contador','Analista'],
    'Nombre':["Pedro","Camila",'Mariano','Federico','Daniel','Graciela','Leandro']
})
```

```python
df_empleados_cobro
```

```python
df_empleados_info
```

#### INNER

```python
df_marge_inner = pd.merge(df_empleados_cobro,df_empleados_info,how="inner",on="ID")
df_marge_inner
```

#### LEFT

```python
df_marge_left = pd.merge(df_empleados_cobro,df_empleados_info,how="left",on="ID")
df_marge_left
```

#### RIGHT

```python
df_marge_right = pd.merge(df_empleados_cobro,df_empleados_info,how="right",on="ID")
df_marge_right
```

#### OUTER

```python
df_marge_outer = pd.merge(df_empleados_cobro,df_empleados_info,how="outer",on="ID")
df_marge_outer.sort_values("ID")
```

## Ejercicio 1.9 --> manejo de valores nulos

```python
dias = pd.date_range(start='2024-04-08', end='2024-04-14')

df_resumen_ventas = pd.DataFrame(
    data = {
        "clientes":[10,np.nan,12,16,20,np.nan,14],
        "pedidos":[4,2,14,10,1,10,12],
        "ingresos":[np.nan,20000,4500,7000,np.nan,5900,12000]
    },
    index=dias)
df_resumen_ventas
```

#### ISNULL --> Vemos que valores son nulos

```python
df_resumen_ventas.isnull()
```

#### ISNULL().SUM() --> suma la cantidad de nulos por cada campo

```python
df_resumen_ventas.isnull().sum()
```

#### DROPNA --> Elimina todas las filas que tengas algun valor nulo o mas de uno

```python
df_resumen_ventas.dropna()
```

#### DROPNA(AXIS=1) --> deja solo las columnas que no tienen ningun valor nulo

```python
df_resumen_ventas.dropna(axis=1)
```

#### FILLNA(0) --> rellena todos los valores nulos con cero. Si se coloca otro numero rrellena con ese numero

```python
df_resumen_ventas.fillna(0)
```

#### FILLNA(DF.MEDIAN()) --> rellena los valores nulos con el valor de la media de la columna

```python
df_resumen_ventas.median()
```

```python
df_resumen_ventas.fillna(df_resumen_ventas.median())
```

#### INTERPOLATE() --> rellena con el promedio del valor anterior y posterior

```python
df_resumen_ventas.interpolate()
```
